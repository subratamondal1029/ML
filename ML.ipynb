```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
# from sklearn.utils import shuffle
import matplotlib.pyplot as pyplot
import pickle
```

```python
data = pd.read_csv('/content/student-mat.csv', sep=';')
data = data[["G1", "G2", "G3", "studytime", "failures", "absences", "freetime"]]
```

```python
predict = "G3"
x = np.array(data.drop(columns=[predict]))
y = np.array(data[predict])
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.1)
score = 0
```

```python
for i in range(20000):
  linear = linear_model.LinearRegression()
  linear.fit(x_train, y_train)
  acc = linear.score(x_test, y_test)

  if score < acc:
    score = acc
    with open("studentmodel.pickle", "wb") as f:
      pickle.dump(linear, f)

print(f"Score: {score}")
linear = pickle.load(open("studentmodel.pickle", "rb"))
```

```python
predictions = linear.predict(x_test)

for x in range(len(predictions)):
  print(round(predictions[x]), f"[{x_test[x][0]}, {x_test[x][1]}]", y_test[x])
```
